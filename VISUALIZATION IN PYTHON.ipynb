```python
#libraries used for visualization are: matplotlib, seaborn, plotly, sweetviz, pywedge and flask

# matplotlib is the base visualization library in python. To use matplotlib,
  you must import it e.g: import matplotlib.pyplot as plt.
        
# pandas uses "plot"
  the pandas lib offers the plot() as a wrapper round the matplotlib lib. The plot()
  allows us to create plots directly on pandas DataFrame.

  plot creates a graph on a DataFrame directly.

# "kind" describes the type of graph e.g: bar chart, scartterd, line, barh(horizontql), histogram

# whatever that is the parenthesis is called an "argument"
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

#                      OOP (Object Oriented Programming )

a method is also a function

# A "attribute" describes an object. A "method" performs action on an object.
#  e.g: df.columns                          df.unique(), x.sort(), x.sum()


# Your variable e.g "df" is alwayas the object
```

```python


df=pd.DataFrame({'name':['john','mary','peter','phil','martins'],
                 'age':[20,30,45,52,60],
                  'gender':['m','f','m','m','m']})
                
```

```python
df
```

```python
# let us create a scatter plot using the panels plot() function.

# Scatter plot is used to ascertain the relationship between two variables
# x-axis contains always the object/name variables while the y-axis contains the numerical values.


df.plot(x='name',y='age',kind='scatter',title='name vs age')
```

```python
df.plot(x='name',y='age',kind='scatter',title='name vs age',color='red')
```

```python
# Creation of a bar chart

df.plot(kind='bar',x='name',y='age',title='name vs age')
```

```python
# Creation of a barh chart


df.plot(kind='barh',x='name',y='age',title='name vs age')
```

```python
# when you do not declare the kind of a chart it will automatically give you a "line chart"

df.plot(x='name',y='age',title='name vs age')
```

```python
df.plot(x='name',y='age',title='name vs age',grid=True)
```

```python
df=pd.DataFrame({'name':['john','mary','peter','phil','martins'],
                 'age':[20,30,45,52,60],
                 'gender':['m','f','m','m','m'],
                 'salary':[500,300,200,100,600]})
                 
```

```python
df
```

```python
df.plot(kind='bar', x='name', y=['age','salary'],title='name vs age-salary')
```

```python
# How to create a box chart 
# it depicts the concentration of the dataset in terms of minimum and maximum
# A box plot is usesd to rep. the relationship between two numerical values. NO object 


df.plot(kind='box', x='salary', y='age',title='salary vs age')
```
